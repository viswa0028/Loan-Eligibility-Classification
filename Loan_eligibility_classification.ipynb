```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tst
```

```python
df= pd.read_csv("/content/drive/MyDrive/Classroom/loan_data_set.csv")
```

```python
df.shape
```

```python
df.head()
```

**Data Preprocessing**

```python
df.drop(['Loan_ID'],axis = 1,inplace = True)
```

```python
df['Gender'].replace({'Male':1,'Female':0},inplace = True)
```

```python
columns = ['Married','Self_Employed']
for column in columns:
  df[column].replace({'Yes':1,'No':0},inplace = True)
```

filling space with mean

```python
df[['Self_Employed','LoanAmount','Loan_Amount_Term','Credit_History']] = df[['Self_Employed','LoanAmount','Loan_Amount_Term','Credit_History']].fillna(df[['Self_Employed','LoanAmount','Loan_Amount_Term','Credit_History']].mean())
```

Creating dummies and concatinating

```python
df = pd.concat([df,pd.get_dummies(df['Property_Area'])],axis = 1)
df.head()
```

```python
df = pd.concat([df,pd.get_dummies(df['Education'])],axis = 1)
df.head()
```

```python
df.drop(['Education','Property_Area'],axis =1,inplace = True)
```

```python
df['Dependents'].replace({'3+':'5'},inplace = True)
```

```python
df['Dependents'] = df['Dependents'].astype(float)
```

```python
df['Loan_Status'].replace({'Y':1,'N':0},inplace  = True)
```

```python
df = df.dropna() #dropping all the empty values
```

**preprocessing using sklearn**

```python
from sklearn.preprocessing import StandardScaler #importing Standardscalar
st = StandardScaler()
df[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']] = st.fit_transform(df[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']])
```

```python
X_train,X_test,y_train,y_test = tst(df.drop(['Loan_Status'],axis = 1),df['Loan_Status'],test_size = 0.2) #splitting the data
```

**Model training**

```python
from sklearn.svm import SVC as svc #importing support vector classifier
```

```python
model = svc()
```

```python
model.fit(X_train,y_train) # model fitting
```

```python
model.score(X_test,y_test) #predicting the outcome
```

You can see that the final outcome accuracy is only 77% but if we use RandomForestClassifier or some other things it will be much higher.We can also change the parameters in the svc to get a better outcome.
